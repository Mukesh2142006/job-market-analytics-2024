```python
# Import all the libraries we need
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

print("All libraries loaded successfully!")
```

```python
# Load the CSV file into Python
# pd.read_csv reads the file and creates a table (called a DataFrame)
df = pd.read_csv('../data/raw/job_postings.csv')

# Print basic info about the dataset
print(f"Dataset loaded successfully!")
print(f"Total rows   : {df.shape[0]:,}")
print(f"Total columns: {df.shape[1]}")
```

```python
# Look at the first 5 rows of data
# This shows us what the data actually looks like
df.head()
```

```python
# This shows every column name and what type of data it holds
# 'object' = text, 'int64' = whole number, 'float64' = decimal number
df.info()
```

```python
# Summary statistics for all number columns
# Shows min, max, average, etc.
df.describe()
```

```python
# Count empty/missing values in each column
missing_count = df.isnull().sum()
missing_pct = (df.isnull().sum() / len(df) * 100).round(2)

missing_df = pd.DataFrame({
    'Missing Count': missing_count,
    'Missing %': missing_pct
}).sort_values('Missing %', ascending=False)

print("Missing values in each column:")
print(missing_df)
```

```python
# Check if any rows are exactly the same
duplicates = df.duplicated().sum()
print(f"Duplicate rows: {duplicates}")
print(f"Duplicate percentage: {(duplicates/len(df)*100):.2f}%")
```

```python
# See all unique experience levels and how many jobs each has
print("Job levels:")
print(df['job_level'].value_counts())
```

```python
# See remote vs onsite vs hybrid breakdown
print("Job types:")
print(df['job_type'].value_counts())
```

```python
# See most common job titles
print("Top 10 job titles:")
print(df['job_title'].value_counts().head(10))
```

```python
# See which companies post the most jobs
print("Top 10 companies:")
print(df['company'].value_counts().head(10))
```

```python
# Top 10 locations
print("Top 10 locations:")
print(df['job_location'].value_counts().head(10))
```
